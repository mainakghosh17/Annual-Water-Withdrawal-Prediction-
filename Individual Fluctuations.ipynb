# Individual Country wise Fluctuations

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.figure(figsize=(10,6))
df = pd.read_csv("total.csv")
df.head(), df.columns
```

# France

```python
plt.plot(df['year'], df['France'], marker='o', linestyle='-', color='blue')
plt.title('France Data Over the Years', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Water withdrawal level', fontsize=12)
plt.show()
```
